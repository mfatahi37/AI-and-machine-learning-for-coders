```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# ساخت مدل
l0 = Dense(units=1, input_shape=[1])
model = Sequential([l0])
model.compile(optimizer='sgd', loss='mean_squared_error')

# داده‌های آموزشی
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

# آموزش مدل
model.fit(xs, ys, epochs=500)

# پیش‌بینی با ورودی جدید
x_test = np.array([[10.0]], dtype=float)
print(model.predict(x_test))

# نمایش وزن‌ها (w و b)
print("Here is what I learned: {}".format(l0.get_weights()))

# اگر می‌خوای دوباره پیش‌بینی کنی، حتما از numpy array استفاده کن
x_test2 = np.array([[10.0]], dtype=float)
print(model.predict(x_test2))
print("Here is what I learned (again): {}".format(l0.get_weights()))
```
